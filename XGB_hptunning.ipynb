```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_Train = scaler.fit_transform(x_Train)
x_Test = scaler.transform(x_Test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    n_estimator = trial.suggest_int('n_estimator', 50 ,150)
    # learning_rate = trial.suggest_float('learning_rate', 0.01 , 0.3)
    max_depth = trial.suggest_int('max_depth', 3,12)
    gamma = trial.suggest_int('gamma', 0 , 5)
    reg_alpha = trial.suggest_int('reg_alpha' ,0 , 5)

    model = xg.XGBRegressor(n_estimator = n_estimator, 
                            max_depth = max_depth , reg_alpha= reg_alpha, random_state = 123 )
    
    score = cross_val_score(model , x_Train, y_Train, cv = 3 , scoring= 'neg_root_mean_squared_error' ).mean()

    return score
```

```python
study = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler())  # We aim to maximize the negative RMSE
study.optimize(objective, n_trials=10)
```

```python
print(f'Best trial accuracy: {study.best_trial.value}')
print(f'Best hyperparameters: {study.best_trial.params}')
```
