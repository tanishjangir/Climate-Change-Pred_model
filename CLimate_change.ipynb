***testing the Random Forest Regressor with random parameters***

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.ensemble import RandomForestRegressor
```

```python
dummy_obj = RandomForestRegressor(n_estimators=50, max_depth=10,n_jobs=-1)
```

```python
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
dummy_obj.fit(x_train,y_train)
```

```python
y_pred = dummy_obj.predict(x_test)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
print("RMSE",root_mean_squared_error(y_test,y_pred))
```

Above score is too bad , lets improve this 

***HYPERPARAMETER TUNNING WITH OPTUNA***
this will find the best parameter values to train with , this technique is better than randomSearch cv and grid search cv(speed+accurracy comaprison)

```python
import optuna
```

```python
def objective(trial):

    n_estimators = trial.suggest_int('n_estimator',100,120)
    max_depth = trial.suggest_int('max_depth',5,15)
    

    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth,n_jobs=-1,random_state=42)

    score = cross_val_score(model,x_train,y_train,cv = 2, scoring='neg_root_mean_squared_error').mean()
    return score
```

```python
study = optuna.create_study(direction='maximize',  sampler=optuna.samplers.TPESampler())
study.optimize(objective,n_trials=10)
```

the hyper parameter tunning or study( in context to Optuna) has been done we can get the best parameters 
a point to note that we tried to maximize the negative mean square  error which is same as reducing the mse 

```python
print("best_value", study.best_trial.value)
```

```python
print("best_value", study.best_params)
```

these are best parameters to train

```python
obj = RandomForestRegressor(n_estimators=103 , max_depth=  15,n_jobs=-1,random_state=42)
```

```python
obj.fit(x_train,y_train)
```

```python
prediction = obj.predict(x_test)
```

```python
print(root_mean_squared_error(y_test,prediction))
```

***you can see the huge improvement***

this is because of optuna hyperparameter tunning 

```python
from sklearn.metrics import r2_score
print("R2 Score:", r2_score(y_test, prediction))
```
